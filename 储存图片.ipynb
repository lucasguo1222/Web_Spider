# 通过爬取图片的url，得到图片的二进制内容然后储存

```python
#User-Agent: Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.80 Safari/537.36
import requests 
#1.发送请求
#请求头
#伪装成浏览器请求的形式， 否则有可能报400 bad request的报错
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36 Edg/87.0.664.66'
}
```

```python
url_list = ['https://p6-tt.byteimg.com/origin/pgc-image/53eeb158296541c09fa04c1421cb9810?from=pc']
```

```python
#2.响应 
### 带参数get请求
### 网址是一张上海的照片
result = []

for url in url_list:
    response = requests.get(url ,headers=headers)
    if response.status_code == 200:
        result.append(response.content)
```

```python
result
```

```python
import time
print(time.time())#获当前时间的时间戳
```

```python
for item in result:
    #将得到的图片保存
    with open(str(time.time())+'.png','wb') as f:
        f.write(item)
        f.close()
```
