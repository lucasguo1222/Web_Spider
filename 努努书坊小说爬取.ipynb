## 爬取历史小说-凯恩舰哗变

```python
import requests 
#1.发送请求
#请求头
#伪装成浏览器请求的形式， 否则有可能报400 bad request的报错
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36 Edg/87.0.664.66'
}
```

```python
#获取所有章节的url：

def get_page_index(html):
    soup = BeautifulSoup(html, 'lxml')
    titles = soup.find(attrs={'cellspacing':'1','cellpadding':'8'}).find_all('td')
    del titles[0]
    lst = []
    for item in titles:
        lst.append(item.a['href'])
    return lst
```

```python
#传入章节的url，返回相应的html
def get_page_detail(url):
    try:
        response = requests.get(url)
        response.encoding='gb2312' #说明编码格式, 但要注意这里有可能不是gb2312
        if response.status_code == 200:
            return response.text #返回一个html格式的响应
        return None
    except RequestException:
        print('请求详情页失败'+url)
        return None
```

```python
#传入每一章节的html， 获取文章内容并存入txt文件
def parse_page_detail(html):
    soup = BeautifulSoup(html, 'lxml')
    title = soup.find('font').string #获取章节标题
    with open('凯恩舰哗变.txt','a+',encoding='gb2312') as file: #将标题写入文档
        file.write('\n\n')
        file.write(title)
    
    #获取段落内容
    paragraph = soup.find('p')
    p = str(paragraph)
    p = p.replace("<br/>", "").replace("<p>", "").replace("</p>", "") #去掉不需要的标签
    print(p)
    with open('凯恩舰哗变.txt','a+',encoding='gb2312') as file: #将内容写入文档
        file.write(p)
```

```python
try:
    response = requests.get(url= 'https://www.kanunu8.com/book/3931/')
    response.encoding='gb2312' #网站编码格式可以通过源代码来查看
    if response.status_code == 200:
        html = response.text
        
except RequestException:
    print('请求索引失败')
```

```python
from bs4 import BeautifulSoup
```

```python
urls = get_page_index(html) #获取所有章节的url
```

```python
for i in urls:
    final_url = 'https://www.kanunu8.com/book/3931/' + i
    print(final_url)
    #response = requests.get(final_url)
    #response.encoding='gb2312' #说明编码格式
    html = get_page_detail(final_url)
    #print(html)
    parse_page_detail(html)
    print('1 chapter finish')
```
